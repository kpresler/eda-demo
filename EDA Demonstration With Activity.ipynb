## Task 2: GDP Growth

```python
# calculate the differences
gdp["changeReal"] = gdp[2007] - gdp[1952] # absolute growth
gdp["changePercent"] = 100*gdp[2007] / gdp[1952] # relative growth
gdp # take a quick look to make sure it seemed to about work
```

```python
# first, look at the largest absolute change
gdp.sort_values(by=["changeReal"], ascending=False)[["changeReal", "changePercent"]]
```
